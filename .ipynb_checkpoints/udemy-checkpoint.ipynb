```python
df7 = pd.read_csv('dataset-2a.csv')
df8 = pd.read_csv('dataset-2b.csv')
df9 = pd.read_csv('dataset-2c.csv')
```

```python
df7
```

```python
df8
```

```python
df9
```

```python
df10 = df7.merge(df8, on='user',how='left')

df10
```

```python
df11 = df10.merge(df9, on='ciudad', how='left')
df11
```

```python
df11.info()
```

```python
df11['FECHA'] = pd.to_datetime(df11['fecha_publicado'], format = "%d/%m/%Y")
```

```python
df11
```

```python
df11['MES'] = df11['fecha_publicado'].apply(lambda x: x[3:])
```

```python
df11
```

```python
graphy =df11.groupby(['MES','ciudad']).agg({'id': 'count'})
graphy2 = graphy.reset_index()
graphy2
```

```python
graphy2.plot.bar(x='ciudad', y='id')
```

```python
df5 = pd.read_csv('titanic.csv')

df5
```

```python
pasajeros = df5.groupby('Sex').agg({'PassengerId': 'count'})
pasajeros
```

```python
pasajeros.plot.bar(y='PassengerId', width=0.4, title='PASAJEROS POR GENERO')
```

```python
import seaborn as sns
```

```python
pasajeros2 = pasajeros.reset_index()

sns.barplot(pasajeros2, x='Sex', y='PassengerId')
```

```python
import matplotlib.pyplot as plt
```
