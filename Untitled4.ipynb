```python
import matplotlib.pyplot as plt
import seaborn as sns


df4 = pd.read_csv('dataset_banco.csv')
```

```python
df4
```

```python
df4.info()
```

```python
df4.dropna(inplace=True)
```

```python
df4.info()
```

```python
categorias = ['job','marital','education','default','housing','loan','contact','month','poutcome','y']

for i in categorias:
    print(f'Columna {i}: {df4[i].nunique()} subniveles')
```

```python
graficas = ['age','balance','day','duration','campaign','pdays','previous']

fig, ax = plt.subplots(nrows=7, ncols=1, figsize=(8,30))
fig.subplots_adjust(hspace=0.5)

for j,col in enumerate(graficas):
    sns.boxplot(x=col, data=df4, ax=ax[j])
    ax[j].set_title(col)
```
